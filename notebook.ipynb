```python
import pandas as pd
import matplotlib.pyplot as plt

netflix_df = pd.read_csv("netflix_data.csv")
```

```python
movies_90s = netflix_df[(netflix_df["type"]=="Movie")&(netflix_df["release_year"] >= 1990) & (netflix_df["release_year"] <= 1999)]
```

```python
netflix_subset=netflix_df[netflix_df["type"]=="Movie"]
subset = netflix_subset[(netflix_subset["release_year"]>=1990)&(netflix_subset["release_year"]<=1999)]


```

```python
movies_dur=movies_90s["duration"]
print(movies_dur)
length_movies=len(movies_dur)
print(length_movies)
import numpy as np
np_arr_dur=np.array(movies_dur)
sorted_arr_dur=np.sort(np_arr_dur)

temp_movie_dur = sorted_arr_dur[0]
print("checking")
print(sorted_arr_dur)
count=1

for i in range(length_movies):
    temp_count=1
    j=i+1 
    while j<length_movies and sorted_arr_dur[j]==sorted_arr_dur[i]:
          temp_count=temp_count+1
          j=j+1
    if temp_count>count:
        count=temp_count
        temp_movie_dur=sorted_arr_dur[j]
    i=i+j-1

duration_exact = temp_movie_dur
print(duration_exact)
```
